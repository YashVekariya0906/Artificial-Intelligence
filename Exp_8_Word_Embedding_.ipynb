```python
sentences = ["India is a beutifull country",
"India is a Powerfull country in the world ",
"Inida is a developing Country",

"Cricket my is favourite game ",
"Sport give us a chance to win the world",

"Technology is the application of scientific knowledge to solve practical problems and improve daily life",
"Technology includes tools, machines, and systems that convert natural resources into usable products"]
```

```python
#IMPORTING BASIC LIBRARIES
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer #TF-BoW
from sklearn.feature_extraction.text import TfidfVectorizer #TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#TF-BoW Embedding
tf_vectorizer=CountVectorizer()
tf_emb=tf_vectorizer.fit_transform(sentences)
```

```python
#similarity matrix using cosine_similarity
tf_sim=cosine_similarity(tf_emb)
```

```python
tf_sim
```

```python
#TF-IDF Embedding
tfidf_vectorizer=TfidfVectorizer()
tfidf_emb=tfidf_vectorizer.fit_transform(sentences)
```

```python
tfidf_sim=cosine_similarity(tfidf_emb)
```

```python
tf_sim
```

```python
# Recommend the most matching/similar sentence
def doc_sim(tfidf_sim,sentences):
  for i in range(len(sentences)):
    sim = tfidf_sim[i].copy()
    sim[i]=-1
    top_index=np.argmax(sim)
    print(sentences[i],"->",sentences[top_index])
```

```python
    doc_sim(tfidf_sim,sentences)
```
