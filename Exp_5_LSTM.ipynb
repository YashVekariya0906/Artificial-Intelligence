```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Input

# Step 1: Create dataset
def create_dataset():
    data = np.array([i for i in range(10)], dtype=float)
    X, y = [], []

    for i in range(len(data) - 3):
        X.append(data[i:i+3])
        y.append(data[i+3])

    return np.array(X), np.array(y)

X, y = create_dataset()

# Normalize data (IMPORTANT)
X = X / 10.0
y = y / 10.0

# Reshape
X = X.reshape((X.shape[0], X.shape[1], 1))

# Step 2: Build model (FIXED warning)
model = Sequential([
    Input(shape=(3, 1)),
    LSTM(50, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Step 3: Train
model.fit(X, y, epochs=300, verbose=0)

# Step 4: Predict
test_input = np.array([7, 8, 9], dtype=float) / 10.0
test_input = test_input.reshape((1, 3, 1))

prediction = model.predict(test_input)

# Convert back to original scale
predicted_value = prediction[0][0] * 10

print("Predicted next value:", predicted_value)
```
