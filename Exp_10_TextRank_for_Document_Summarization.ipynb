```python
import numpy as np
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer

text = """Nature gives us a sense of peace and helps reduce stress in our daily lives.
The sound of birds and the beauty of flowers make the environment pleasant and refreshing."""

sentences = text.split(". ")
sentences = [s.strip() for s in sentences if s]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(sentences)
similarity_matrix = (X * X.T).toarray()

# Step 5: Build graph
nx_graph = nx.from_numpy_array(similarity_matrix)
scores = nx.pagerank(nx_graph)
ranked_sentences = sorted(((scores[i], s) for i, s in enumerate(sentences)), reverse=True)
print("Summary:")
for score, sentence in ranked_sentences[:2]:
    print("-", sentence)
```
